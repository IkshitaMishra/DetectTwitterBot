```python
import sys, codecs, json
from twython import TwythonStreamer, Twython
from datetime import datetime
from time import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in OAuth Tokens
app_key = "key"
app_sec = "sec"
user_key = "user-key"
user_sec = "user-sec"
```

Filter API Class

```python
tweets_data = []
class StreamAPI(TwythonStreamer):
    def on_success(self, data):
        dt = datetime.now()
        filename = "data/twitterdata-%s.txt" % (dt.strftime("%Y-%m-%d"),)
        if 'text' in data:
            print(data['text'])
            fp = codecs.open(filename, 'a', 'utf-8')
            fp.write(json.dumps(data))
            fp.write("\n")
            tweets_data.append(data)
            fp.close()

    def on_error(self, status_code, data):
        print("ERROR!")
        print(status_code)
```

```python
stream = StreamAPI(app_key, app_sec, user_key, user_sec)
```

```python
stream.statuses.filter(locations='-122.75,36.8,-121.75,37.8') #Bounding box for San Fransisco
```
